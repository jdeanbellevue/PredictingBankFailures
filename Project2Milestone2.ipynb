```python
missing_values_balance_sheet = balance_sheet.isnull().sum()

balance_sheet['Failure Date'] = pd.to_datetime(balance_sheet['Failure Date'], errors='coerce')

duplicates_balance_sheet = balance_sheet.duplicated().sum()

column_names_issues_balance_sheet = balance_sheet.columns != balance_sheet.columns.str.strip()

{
    "missing_values_balance_sheet": missing_values_balance_sheet[missing_values_balance_sheet > 0],
    "duplicates_balance_sheet": duplicates_balance_sheet,
    "column_names_issues_balance_sheet": column_names_issues_balance_sheet.any(),
    "data_types_balance_sheet": balance_sheet.dtypes
}
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

rows_with_missing_values_balance_sheet = balance_sheet[balance_sheet.isnull().any(axis=1)]

numerical_columns_balance_sheet = balance_sheet.select_dtypes(include=['float64'])

plt.figure(figsize=(15, 10))
sns.boxplot(data=numerical_columns_balance_sheet)
plt.xticks(rotation=90)
plt.title('Boxplots of Numerical Columns in balance_sheet Dataset')
plt.show()

```

```python
balance_sheet_cleaned = balance_sheet.drop(columns=['Total Liabilities(2)'])

rows_with_missing_values_cleaned = balance_sheet_cleaned[balance_sheet_cleaned.isnull().any(axis=1)]

balance_sheet_cleaned.head()
```

```python
balance_sheet_cleaned = balance_sheet_cleaned.dropna()

non_negative_columns = balance_sheet_cleaned.select_dtypes(include=['float64']).columns  
negative_values = (balance_sheet_cleaned[non_negative_columns] < 0).sum()

{
    "balance_sheet_cleaned_head": balance_sheet_cleaned.head(),
    "balance_sheet_cleaned_shape": balance_sheet_cleaned.shape,
    "negative_values": negative_values[negative_values > 0]
}
```

```python
balance_sheet_cleaned = balance_sheet_cleaned.drop(columns=['Assets in Liquidation', 
                                                            'Estimated Loss on Assets in Liquidation(1)'])

balance_sheet_cleaned['Due from FDIC Corp and Receivables'] = balance_sheet_cleaned[
    'Due from FDIC Corp and Receivables'
].apply(lambda x: max(0, x))

balance_sheet_cleaned.head() 
```

```python
key_numerical_columns = [
    'Cash and Investments', 'Total Assets', 
    'Administrative Liabilities', 'FDIC Subrogated Deposit Claim',
    'Net Worth (Deficit)'
]

plt.figure(figsize=(15, 10))
balance_sheet_cleaned[key_numerical_columns].hist(bins=15, figsize=(15, 10), layout=(2, 3))
plt.suptitle('Histograms of Key Numerical Variables')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

correlation_matrix = balance_sheet_cleaned[key_numerical_columns].corr()

plt.figure(figsize=(10, 7))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Correlation Matrix of Key Numerical Variables')
plt.show()

descriptive_stats = balance_sheet_cleaned[key_numerical_columns].describe()

descriptive_stats
```

```python
plt.figure(figsize=(15, 7))
plt.subplot(2, 1, 1)
sns.lineplot(data=balance_sheet_cleaned, x='Failure Date', y='Total Assets')
plt.title('Total Assets over Time')
plt.ylabel('Total Assets')

plt.subplot(2, 1, 2)
sns.lineplot(data=balance_sheet_cleaned, x='Failure Date', y='Net Worth (Deficit)')
plt.title('Net Worth (Deficit) over Time')
plt.ylabel('Net Worth (Deficit)')

plt.tight_layout()
plt.show()

receivership_count = balance_sheet_cleaned['Receivership'].value_counts()

plt.figure(figsize=(15, 7))
plt.subplot(2, 1, 1)
sns.boxplot(data=balance_sheet_cleaned, x='Receivership', y='Total Assets')
plt.title('Total Assets by Receivership')
plt.subplot(2, 1, 2)
sns.boxplot(data=balance_sheet_cleaned, x='Receivership', y='Net Worth (Deficit)')
plt.title('Net Worth (Deficit) by Receivership')
plt.tight_layout()
plt.show()

{
    "receivership_count": receivership_count
}
```

```python
stats_total_assets = balance_sheet_cleaned['Total Assets'].describe()
stats_net_worth = balance_sheet_cleaned['Net Worth (Deficit)'].describe()

IQR_total_assets = stats_total_assets['75%'] - stats_total_assets['25%']
outlier_threshold_high_total_assets = stats_total_assets['75%'] + 1.5 * IQR_total_assets
outlier_threshold_low_total_assets = stats_total_assets['25%'] - 1.5 * IQR_total_assets

IQR_net_worth = stats_net_worth['75%'] - stats_net_worth['25%']
outlier_threshold_high_net_worth = stats_net_worth['75%'] + 1.5 * IQR_net_worth
outlier_threshold_low_net_worth = stats_net_worth['25%'] - 1.5 * IQR_net_worth

outliers_total_assets = balance_sheet_cleaned[
    (balance_sheet_cleaned['Total Assets'] > outlier_threshold_high_total_assets) |
    (balance_sheet_cleaned['Total Assets'] < outlier_threshold_low_total_assets)
]

outliers_net_worth = balance_sheet_cleaned[
    (balance_sheet_cleaned['Net Worth (Deficit)'] > outlier_threshold_high_net_worth) |
    (balance_sheet_cleaned['Net Worth (Deficit)'] < outlier_threshold_low_net_worth)
]

{
    "outliers_total_assets": outliers_total_assets[['Receivership', 'Total Assets', 'Failure Date']],
    "outliers_net_worth": outliers_net_worth[['Receivership', 'Net Worth (Deficit)', 'Failure Date']]
}
```
